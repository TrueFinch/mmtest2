```python
import numpy as np
from matplotlib import pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller, acf
import statsmodels.api as sm
import pandas as pd
from scipy import stats
import pca
```

Загрузим исходные данные и посмотрим их на графике.

```python
s = '5400 5630 5420 5680 5400 5620 5410 5700 5590 5700 5680 5700 5600 5680 5400 5570 5410 5500 5410 5510 5490 5800 5530 5700 5600 5700 5550 5630 5600 5700 5680 5700 5520 5530 5380 5600 5580 5700 5620 5730 5680 5800 5540 5640 5610 5720 5610 5700 5900 5990 5900 5930 5940 6300 5930 6020 6000 6200 6300 6270'
data = np.array(list(map(int, s.split(' '))))
plt.plot(range(60), data)
```

Построим функцию ACF

```python
plot_acf(data)
acf(data)
```

По графику ACF характерен для нестационарных рядов, чтобы в этом убедиться посмотрим на результаты теста Дикки-Фуллера

```python
adfuller(data)
```

Как видим ADF статистика больше, чем все приведенные критические значения, а также p-value велико, то можно утверждать, что ряд не стационарен

Для удобства приведем данные немного к другому виду

```python
rng = pd.date_range('2015-02-24', periods=60, freq='D')
df = pd.DataFrame({'Val': data }, index=rng) 
```

По виду полученного ряда можно сказать, что он описывается аддитивной моделью (магнитуда не возрастает), разложим исходный ряд в соответствии с выбранной моделью

```python
decompose = sm.tsa.seasonal_decompose(df, model='additive')
trend = decompose.trend
season = decompose.seasonal
res = decompose.resid
decompose.plot()
```

Полученный тренд хорошо отражает наши данные визуально.

```python
plt.plot(range(60), trend.values)
plt.plot(range(60), data)
```

Средняя ошибка аппроксимации мала

```python
abs(((data - trend) / data)).sum() / len(data)
```

Также мал линейный коэффициент корреляции

```python
((data - data.mean()) / data.std()).sum() * ((trend - trend.mean()) / trend.std()).sum() / (len(data) - 1)
```

F-критерий и t-критерий также подтверждают нашу гипотезу

```python
#F-критерий
alpha = 0.05
F = trend.var() / data.var()
df1 = len(trend) - 1; df2 = len(data) - 1
p_value = stats.f.cdf(F, df1, df2)
print(p_value)
if alpha < p_value:
    print('reject')
else:
    print('accept')
```

```python
#t-критерий
stats.ttest_ind(a=trend.dropna(), b=data, equal_var=True) 
#p_value > 0.05
```

Посчиатем ошибку модели

```python
#mean abs error
(abs(res)).sum() / len(data)
```

```python
#mean deriv
(trend + season - data.mean()).mean()
```

Посмотрим на критерий AIC с различными параметрами, resid достигает

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
import itertools

p = d = q = range(0,3)
pdq = list(itertools.product(p,d,q))
combs = {}
aics = []

for combination in pdq:
    try:
        model = SARIMAX(diff2, order=combination, seasonal=False)
        model = model.fit()
        combs.update({model.aic : combination}) 
        aics.append(model.aic)
    except:
        continue
best_aic = min(aics)
print(combs[best_aic])
model = SARIMAX(diff2, order=combs[best_aic], seasonal=False)
model1 = model.fit()
model1.resid.plot(kind='kde')
```

Ошибка полученнного решения

```python
(abs(model1.resid)).sum() / len(data)
```

```python
diff2.plot()
model1.fittedvalues.plot()
```

```python
combs = {}
bics = []

for combination in pdq:
    try:
        model = SARIMAX(diff2, order=combination, seasonal=False)
        model = model.fit()
        combs.update({model.bic : combination})
        bics.append(model.bic)
    except:
        continue
best_bic = min(bics)
model = SARIMAX(diff2, order=combs[best_bic], seasonal=False)
model2 = model.fit()
model2.resid.plot(kind='kde')
```

Ошибка полученнного решения немного больше, чем для AIC

```python
(abs(model2.resid)).sum() / len(data)
```

Визуально, также заметно модель подобранная с помощью AIC немного лучше описывает данные

```python
diff2.plot()
model2.fittedvalues.plot()
model1.fittedvalues.plot()
```

Центрирование и нормирование данных проводится по формуле \
$ X_{1}=\frac{x_{1 j}-\bar{x}_{1}}{\hat{s}_{1}} ; X_{2}=\frac{x_{2 j}-\bar{x}_{2}}{\hat{s}_{2}} ; \ldots ; X_{m}=\frac{x_{m j}-\bar{x}_{m}}{\hat{s}_{m}} ; j=\overline{1, n} $

```python
df -= df.mean()
df /= df.std()
df
```

Ковариационная матрица вычисляется по формуле \
$ \Sigma^{(m)}=\left(\begin{array}{cccc}
\operatorname{cov}\left(X_{1}, X_{1}\right) & \operatorname{cov}\left(X_{1}, X_{2}\right) & \ldots & \operatorname{cov}\left(X_{1}, X_{m}\right) \\
\operatorname{cov}\left(X_{2}, X_{1}\right) & \operatorname{cov}\left(X_{2}, X_{2}\right) & \ldots & \operatorname{cov}\left(X_{2}, X_{m}\right) \\
\ldots & \ldots & \ldots & \ldots \\
\operatorname{cov}\left(X_{m}, X_{1}\right) & \operatorname{cov}\left(X_{m}, X_{2}\right) & \ldots & \operatorname{cov}\left(X_{m}, X_{m}\right)
\end{array}\right) $ \
Можно использовать формулу \
$ \Sigma=\frac{1}{n-1} X \cdot X^{T} $

```python
covmat = df.cov()
covmat
```

Для нахожденя собственных значений решим характеристическое уравнение  
$ \left|\sum-\lambda E\right|=0 $

```python
vals, vecs = np.linalg.eig(covmat.to_numpy())
', '.join(map(str, vals))
```

Находим собственный вектор $ v_{i} $ для каждого собственного значения $ \lambda_{i} $ с помощью системы линейных однородных уравнений
$ \left(\sum-\lambda_{i} E\right) v_{i}=0 $ \
При этом должно соблюдаться условие: \
$ \left|v_{i}\right|=1 $ \
или \
$ \sum_{j=1}^{n} v_{i j}^{2}=v_{i} \cdot v_{i}^{T}=1 $ \
т.е. собственные векторы необходимо нормировать. \
Далее составим ортогональную матрицу собственных векторов \
$ V=\left(\begin{array}{llll}
v_{1} & v_{2} & \ldots & v_{m}
\end{array}\right)=\left(\begin{array}{cccc}
v_{11} & v_{21} & \ldots & v_{m 1} \\
v_{12} & v_{22} & \ldots & v_{m 2} \\
\ldots & \ldots & \ldots & \ldots \\
v_{1 n} & v_{2 n} & \ldots & v_{m n}
\end{array}\right) $

```python
vecs = pd.DataFrame(vecs)
-vecs
```

Информативность главных компонент вычисляется по формуле \
$ I_{p}=\frac{\lambda_{1}+\ldots+\lambda_{p}}{D x_{1}+\ldots+D x_{m}}, p=1,2, \ldots, m $

```python
vecs = vecs.to_numpy()
narr = df.to_numpy()
I = np.zeros(len(vals))
sum_vals, sum_x = 0, 0
for i in range(len(vals)):
    sum_vals += vals[i]
    sum_x += (narr.std(axis=0) ** 2)[i]
    I[i] = sum_vals / sum_x
pd.DataFrame(I)
```

Меньше всего информацию в себе несет первая переменная, за ней вторая

Новые признаки вычисляются по формуле \
$ Z=L^{T} X $ \
где \
$ \begin{gathered}
z_{i}=v_{i} \cdot X \\
z_{i}=v_{i}^{T} X=v_{i 1} X_{1}+v_{i 2} X_{2}+\ldots+v_{i m} X_{m} ; i=1, \ldots, k
\end{gathered} $

```python
v = -vecs[:, [0, 2]]
Xnew = np.dot(df.to_numpy(), v)
pd.DataFrame(Xnew)
```
